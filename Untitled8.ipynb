```python
# To keep the page organized, do all imports here
from sqlalchemy import create_engine, text
import pandas as pd

# Matplotlib for additional customization
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns 
sns.set()

# Suppress warnings (warnings do not prevent the code from executing)
import warnings
warnings.filterwarnings("ignore")

# Database credentials
postgres_user = 'dabc_student'
postgres_pw = '7*.8G9QH21'
postgres_host = '142.93.121.174'
postgres_port = '5432'
postgres_db = 'houseprices'

# Use the credentials to start a connection
engine = create_engine('postgresql://{}:{}@{}:{}/{}'.format(
    postgres_user, postgres_pw, postgres_host, postgres_port, postgres_db))
prices_df = pd.read_sql_table('houseprices', con=engine.connect())

# Remove the connection
engine.dispose()
```

```python
sns.distplot(prices_df['saleprice'])
```

```python
ladata_filtered = prices_df.loc[prices_df['lotarea'] <=50000, 'lotarea']

plt.boxplot(ladata_filtered)
plt.title('Boxplot of Lot Area')
plt.ylabel('Lot Area (sq ft)')
```

```python
sns.boxplot(x=prices_df['overallcond'], y=prices_df['saleprice'])
plt.title('Boxplot of Sale Price by Overall Condition')
plt.xlabel('Overall Condition')
plt.ylabel('Sale Price')
```

```python
sns.boxplot(x=prices_df['overallcond'], hue= prices_df['centralair'], y=prices_df['saleprice'])
plt.title('Boxplot of Sale Price by Overall Condition')
plt.xlabel('Overall Condition')
plt.ylabel('Sale Price')
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(y="lotarea", x="saleprice", data=prices_df)
plt.ylabel("Lot Area")
plt.xlabel("Sale Price")
plt.title('Lot Area to Sale Price Relationship')
plt.show()
```
